```python
from sclibrary import dataset_loader, SCPlot

sc, coordinates = dataset_loader.load_paper_data()

scplot = SCPlot(simplical_complex=sc, coordinates=coordinates)
```

```python
from matplotlib import pyplot as plt

fig, _ = plt.subplots(figsize=(10, 10))

edge_flow = [0, 1, 1, 1, 2, 4, 9, 2, 3, 1]
edges = [(0, 1),
 (0, 2),
 (0, 3),
 (1, 2),
 (2, 3),
 (2, 5),
 (3, 4),
 (4, 5),
 (4, 6),
 (5, 6)]


edge_flow = {
    edge: flow
    for edge, flow in zip(edges, edge_flow)
}

scplot.draw_sc_nodes(
    node_size=1000,
    font_size=25,
    with_labels=True
)
scplot.draw_sc_edges(
    edge_flow=edge_flow,
    edge_width=5,
    arrowsize=30
    
)

scplot.draw_edge_labels(
    edge_labels=edge_flow,
    font_size=20
)
```
