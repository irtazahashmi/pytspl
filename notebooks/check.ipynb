```python
import pandas as pd
import numpy as np
```

```python
B1_filename = "data/test_dataset/B1_chicago_sketch.csv"
#B1_filename = "data/blah/B1.csv"

B1 = pd.read_csv(B1_filename, header=None).to_numpy()

def extract_edges(B1):
    num_edges = B1.shape[1]
    
    nodes = set()
    edges = []
    
    for j in range(num_edges):
        col = B1[:,j]
        from_node = np.where(col == -1)[0][0]
        to_node = np.where(col == 1)[0][0]
        
        nodes.add(from_node)
        nodes.add(to_node)

        edges.append((from_node, to_node))

    # sort
    nodes = list(nodes)
    nodes.sort()
    edges.sort()
    edges = [tuple(edge) for edge in edges]
    return nodes, edges

nodes, edges = extract_edges(B1)
print(len(edges))
edges
```

```python
B2_filename = "data/test_dataset/B2t_chicago_sketch.csv"
#B2_filename = "data/blah/B2.csv"

B2 = pd.read_csv(B2_filename, header=None).to_numpy().T

num_edges = B2.shape[0]
num_triangles = B2.shape[1]
print(num_edges, num_triangles)

# Function to extract triangles and their edges
def extract_triangles(B2):
    _, edges = extract_edges(B1)
    edges = np.array(edges)
    
    triangles = []
    for j in range(num_triangles):
        print(f"Triangle {j}")
        # Check each column of B2 for triangles
        col = B2[:, j]
        print(col)
        # get pos of 1s
        ones = np.where(col != 0)[0]
        print(ones)
        triangle = edges[ones]
        print(triangle)
        triangle = tuple(set(triangle.flatten()))
        # sort 
        triangle = tuple(sorted(triangle))
        triangles.append(triangle)
        print(triangle)
        print("-" * 50)
    
    return triangles


triangles = extract_triangles(B2)
print(len(triangles))
triangles
```

```python
from sclibrary.io.network_reader import read_B1

g = read_B1(B1_filename=B1_filename)
sc = g.to_simplicial_complex(triangles=triangles)

print("nodes:", len(sc.nodes))
print("edges:", len(sc.edges))
print("triangles:", len(sc.triangles))
```
