```python
from pytspl import load_dataset, SCPlot

sc, coordinates, flow = load_dataset("paper")

# create a plot
scplot = SCPlot(sc, coordinates)
```

```python
import matplotlib.pyplot as plt
fig, _ = plt.subplots(figsize=(8, 10))

scplot.draw_sc_nodes(
    node_size=1000,
    font_size=25,
    with_labels=True
)
scplot.draw_sc_edges(
    #edge_flow=flow,
    edge_width=10,
    arrowsize=30,
    edge_cmap=plt.cm.Blues,   
)

# scplot.draw_edge_labels(
#     edge_labels=flow,
#     font_size=20
# )
```
